## Data Cleaning

```python
# Mencari missing value
data.isnull().sum()
```

```python
# total baris duplikat
print("Total baris duplikat:", data.duplicated().sum())
```

```python
# mapping data kategorikal

print("Attack Type unique values:")
print(data["Attack Type"].unique(), "\n")

print("Target Industry unique values:")
print(data["Target Industry"].unique(), "\n")

print("Attack Source unique values:")
print(data["Attack Source"].unique(), "\n")

print("Security Vulnerability Type unique values:")
print(data["Security Vulnerability Type"].unique(), "\n")

print("Defense Mechanism Used unique values:")
print(data["Defense Mechanism Used"].unique(), "\n")
```

```python
# Attack Type mapping
attack_type_mapping = {
    'Phishing': 0,
    'Ransomware': 1,
    'Man-in-the-Middle': 2,
    'DDoS': 3,
    'SQL Injection': 4,
    'Malware': 5
}

# Target Industry mapping
target_industry_mapping = {
    'Education': 0,
    'Retail': 1,
    'IT': 2,
    'Telecommunications': 3,
    'Government': 4,
    'Banking': 5,
    'Healthcare': 6
}

# Attack Source mapping
attack_source_mapping = {
    'Hacker Group': 0,
    'Nation-state': 1,
    'Insider': 2,
    'Unknown': 3
}

# Security Vulnerability Type mapping
vulnerability_mapping = {
    'Unpatched Software': 0,
    'Weak Passwords': 1,
    'Social Engineering': 2,
    'Zero-day': 3
}

# Defense Mechanism Used mapping
defense_mapping = {
    'VPN': 0,
    'Firewall': 1,
    'AI-based Detection': 2,
    'Antivirus': 3,
    'Encryption': 4
}

# Apply mapping ke DataFrame
data["Attack Type_encoded"] = data["Attack Type"].map(attack_type_mapping)
data["Target Industry_encoded"] = data["Target Industry"].map(target_industry_mapping)
data["Attack Source_encoded"] = data["Attack Source"].map(attack_source_mapping)
data["Vulnerability Type_encoded"] = data["Security Vulnerability Type"].map(vulnerability_mapping)
data["Defense Mechanism_encoded"] = data["Defense Mechanism Used"].map(defense_mapping)

data.head()
```

# B. Analisis Outlier

## Outlier

```python
plt.figure(figsize=(8,4))
sns.boxplot(x=data["Attack Type_encoded"])
plt.title("Persebaran Data Attack Type")
plt.show()
```

```python

mil = "Financial Loss (in Million $)"

mean = data[mil].mean()
std = data[mil].std()

upper_limit = mean + 2*std
lower_limit = mean - 2*std

print(f"Column: {mil}")
print(f"Mean: {mean:.2f}, Std: {std:.2f}")
print(f"Upper limit: {upper_limit:.2f}, Lower limit: {lower_limit:.2f}")

# Ambil outlier
outliers = data[(data[mil] > upper_limit) | (data[mil] < lower_limit)]

print("\nNormal data count:", len(data) - len(outliers))
print("Outliers count (Mean ± 2*Std):", len(outliers))
```

```python
data_clean = pd.read_csv("Global_Cybersecurity_Threats_2015-2024.csv")

Q1 = data_clean["Financial Loss (in Million $)"].quantile(0.25)
Q2 = data_clean["Financial Loss (in Million $)"].quantile(0.5)
Q3 = data_clean["Financial Loss (in Million $)"].quantile(0.75)

print("Q1:", Q1)
print("Q3:", Q3)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)

outliers_iqr = data_clean[
    (data_clean["Financial Loss (in Million $)"] < lower_bound) |
    (data_clean["Financial Loss (in Million $)"] > upper_bound)
]

print("\nNormal:")
print(len(data_clean) - len(outliers_iqr))

print("\nOutliers (IQR):")
print(len(outliers_iqr))

```

```python
plt.hist(data_clean["Financial Loss (in Million $)"], bins=30, edgecolor='black', alpha=0.5, color='green')

# Tambahkan garis vertikal pada standar deviasi
plt.axvline(x=Q2, color='red', linestyle='dashed', linewidth=1, label=f'Mean = {mean}')
plt.axvline(x=lower_bound, color='blue', linestyle='dashed', linewidth=1, label=f'{lower_bound:.2f}')
plt.axvline(x=upper_bound, color='blue', linestyle='dashed', linewidth=1, label=f'{upper_bound:.2f}')

# Tambahkan label dan judul
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title(f'IQR = {IQR:.2f}')

# Tambahkan legenda
plt.legend()

# Tampilkan plot
plt.show()
```

```python
data_normal = data_clean[
    ((data_clean["Financial Loss (in Million $)"] > lower_limit) & (data_clean["Financial Loss (in Million $)"] < upper_limit))
]
data_normal.head()
```

# Summary

## Insight dan Kesimpulan Awal

--Pola Menarik dari Data--

- Serangan dengan jenis tertentu (misalnya DDoS/Phishing) cenderung lebih sering terjadi.

- Industri tertentu menjadi target utama (misalnya finance/healthcare)

- Jumlah pengguna terdampak tidak selalu linier dengan kerugian finansial


--Variabel Paling Berpengaruh--
- Financial Loss dan Number of Affected Users

- Attack Type dan Target Industry

- Defense Mechanism Used
- Incident Resolution Time

--Potensi Masalah--
- Imbalance data: beberapa kategori (Attack Type / Industry) mungkin terlalu dominan bisa memengaruhi modeling klasifikasi.

- Multicollinearity: kemungkinan ada korelasi tinggi antara variabel numerik, misalnya Financial Loss dengan Affected Users.

- Encoding kategori: beberapa variabel kategorikal dengan banyak level perlu ditangani

- Temporal bias: jika data per tahun tidak seimbang, bisa ada bias waktu.

--Pertanyaan Lanjutan untuk Modeling--
- Apakah jenis serangan tertentu (Attack Type) hanya merugikan finansial, atau juga berdampak pada jumlah user yang terdampak secara signifikan?

- Apakah mekanisme pertahanan tertentu (contoh: AI-based Detection vs Firewall) lebih efektif terhadap serangan spesifik (misalnya DDoS vs Phishing), bukan sekadar rata-rata?

- Apakah pola serangan dari “Nation-state” berbeda dengan “Hacker Group” dalam target industri yang disasar?

- Apakah negara tertentu hanya “tampak” aman karena jumlah serangan rendah, padahal jika dihitung kerugian per serangan, lebih parah?

## kesimpulan
- Financial Loss dan Number of Affected Users adalah variabel paling penting untuk menilai dampak serangan.

- Incident Resolution Time berkorelasi dengan mekanisme pertahanan semakin baik defense mechanism, semakin cepat penyelesaian.

- Attack Type dan Target Industry jadi pembeda utama pola serangan, sedangkan Attack Source memberi konteks tambahan.

- Defense Mechanism ada pengaruh, tapi tidak selalu konsisten membatasi kerugian finansial.

- Perlu eksplorasi fitur tambahan (misalnya tren tahunan, kombinasi Attack Type + Vulnerability) untuk pemodelan yang lebih tajam.
